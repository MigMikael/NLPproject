```python
def vocab_size(text):
    return len(text.split(" "))
```

```python
def percent(word, text):
    occurs_count = text.count(word)
    size = vocab_size(text)
    return (occurs_count / size) * 100

the_text = "to be or not to be"
print("vocab size : ", vocab_size(the_text))
print("percent of 'to' : ", percent("to", the_text))
```
